```python
## Ben Robin's 5.16 two dimensional lists 

a = [[77,68,86,73], [96,87,89,81], [70,90,86,81]]
            
for row in a:
    for item in row:
        print(item, end=" ")
    print()
```

```python
for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f'a[{i}][{j}]={item} ', end=' ')
    print()
```

```python
# Ben Robin Check 5.16 Given the two-by-three integer list t
t = [[10, 7, 3], [20, 4, 17]]
# Determinte and display avg. of t's elements using nested for statements to iterate through the elements 

total = 0 

items = 0 

for row in t:
    for item in row:
        total += item
        items += 1 
total/items
```

```python
total = 0 
items = 0 

for row in t:
    total += sum(row)
    items += len(row)
print({total}, 'sum of items', {items}, 'count of items')
```
